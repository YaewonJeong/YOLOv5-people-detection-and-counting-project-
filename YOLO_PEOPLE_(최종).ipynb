```python
from glob import glob 
import os 
train_label_list = glob('/content/yolov5/train/labels/*.txt')
for path in train_label_list:
  print(path)
  f = open(path, "rt") 
  person = []
  for line in f:
    if line.startswith("14"):
      person.append(line)
  f.close()
  if len(person) == 0 : # person 이 없으면 삭제한다 
    os.remove(path)
    os.remove(path.replace(".txt", ".jpg").replace("/labels/","/images/"))
  else : 
    f2 = open(path, 'wt')
    for p in person:
      f2.write(p.replace("14","0",1))
    f2.close()
```

```python
from glob import glob 
import os 
train_label_list = glob('/content/yolov5/valid/labels/*.txt')
for path in train_label_list:
  print(path)
  f = open(path, "rt") 
  person = []
  for line in f:
    if line.startswith("14"):
      person.append(line)
  f.close()
  if len(person) == 0 : # person 이 없으면 삭제한다 
    os.remove(path)
    os.remove(path.replace(".txt", ".jpg").replace("/labels/","/images/"))
  else : 
    f2 = open(path, 'wt')
    for p in person:
      f2.write(p.replace("14","0",1))
    f2.close()
```

```python
from glob import glob

train_img_list = glob('/content/yolov5/train/images/*.jpg')
val_img_list = glob('/content/yolov5/valid/images/*.jpg')

print(len(train_img_list), len(val_img_list))
```

```python
with open('/content/yolov5/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/yolov5/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')

import yaml

with open('/content/yolov5/data.yaml', 'r') as f:
  data = yaml.safe_load(f)

print(data)

data['train'] = '/content/yolov5/train.txt'
data['val'] = '/content/yolov5/val.txt'

with open('/content/yolov5/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
